```python
import numpy as np
import pandas as pd
import folium
```

```python
df = pd.read_csv('pokemon-spawns.csv')
poke4000 = df.head(1000)
poke4000
```

```python
a = np.random.randn()/150
b = np.random.randn()/15
lat = 37.320000 + a
lng = -122.000000 + b
m = folium.Map([lat, lng], tiles ='OpenStreetMap', zoom_start = 15)
for i in range(0, 1000):
    c = np.random.randn()/100
    d = np.random.randn()/100
    hai = poke4000.loc[i]['name']
    depan = "https://raw.githubusercontent.com/PokeAPI/sprites/master/sprites/pokemon/"
    isi = str(poke4000.loc[i]['num'])
    belakang = ".png"
    pokemon = depan+isi+belakang
    gambar = '<img width= "250px" src="'+pokemon+'"/>'+'<center>'+hai+'</center>'
    Icon = folium.features.CustomIcon(pokemon, icon_size=(60,60))
    folium.Marker([poke4000.loc[i]['lat']+c, poke4000.loc[i]['lng']+d], icon=Icon, tooltip = gambar, popup=hai).add_to(m)
m
```
